## **<font color='blue'>Voting Classifiers**

A voting classifier is a machine learning technique that combines the predictions of multiple individual classifiers to make a final decision or prediction. It leverages the wisdom of the crowd by allowing each classifier to "vote" for the most likely class label, and the class label with the majority of votes is chosen as the final output.

### **Types of Voting Classifiers**

There are two main types of voting classifiers:

**Hard Voting:** In hard voting, each individual classifier's prediction contributes equally to the final decision. The class label that receives the most votes becomes the overall prediction.

**Soft Voting:** Soft voting takes into account the class probabilities predicted by each classifier. The class probabilities are averaged across all classifiers, and the class with the highest cumulative probability is selected as the final prediction.

### **How Voting Classifier Works**

This ensemble technique leverages the principle of combining diverse models to improve overall accuracy and robustness. Here's how a voting classifier works:

1. **Selection of Individual Classifiers:** Choose a set of diverse individual classifiers, each with its own strengths, weaknesses, and unique perspectives. These could be different algorithms or variations of the same algorithm with different parameter settings.

2. **Training the Individual Classifiers:** Train each individual classifier on the same training dataset. Each classifier learns to make predictions based on the input features and associated labels.

3. **Prediction Phase:** During the prediction phase, when new data points are presented, each individual classifier makes its own prediction based on its learned rules.
For classification tasks, each classifier predicts a class label for the new data point.
For regression tasks, each classifier predicts a continuous value.

4. **Voting Strategy:** In hard voting, the class label that receives the majority of votes from individual classifiers is chosen as the final prediction. It's like a "majority rules" approach.
In soft voting, the class probabilities predicted by each classifier are averaged across all classifiers. The class with the highest cumulative probability becomes the final prediction.

5. **Final Prediction:** For classification tasks, the class label determined by the chosen voting strategy is the final predicted class.
For regression tasks, the averaged or combined predictions from individual classifiers are used as the final predicted value.

## **<font color='blue'>Light GBM**

Light Gradient Boosting Machine (LightGBM) is an advanced and efficient gradient boosting framework designed for high-performance machine learning tasks. It is specifically optimized for speed, memory usage, and handling large datasets. LightGBM leverages novel techniques to enhance the training process and improve predictive accuracy. Here's a comprehensive explanation of LightGBM:

1. **Histogram-Based Approach:** LightGBM employs a histogram-based technique for finding the best splits during the construction of decision trees. This approach reduces memory usage and speeds up training, making it suitable for large datasets.

2. **Leaf-Wise Tree Growth:** LightGBM uses a leaf-wise tree growth strategy, where it selects the leaf with the maximum reduction in the loss function for splitting. This can lead to more accurate and deeper trees compared to other gradient boosting implementations.

3. **Categorical Feature Handling:** LightGBM efficiently handles categorical features by choosing optimal split points for them. This eliminates the need for extensive preprocessing like one-hot encoding.

4. **Data Parallelism:** LightGBM supports data parallelism, allowing it to effectively use multiple CPU cores for training. This enhances its speed and scalability, making it well-suited for large-scale tasks.

5. **Gradient-Based One-Side Sampling (GOSS):** LightGBM incorporates GOSS, a sampling method that prioritizes instances with larger gradients during training. This helps improve the model's generalization ability and contributes to faster convergence.

6. **Regularization Techniques:** LightGBM offers L1 and L2 regularization to prevent overfitting and enhance model stability.

7. **Hyperparameter Tuning:** LightGBM provides a range of hyperparameters that can be tuned to optimize model performance and prevent overfitting.

### **How LightGBM Works**


1. **Data Input:** Accepts tabular data with features and target variables.

2. **Initialization:** Begins with a simple initial model.

3. **Gradient-Based Learning:** Optimizes predictions using gradients of the loss function.

4. **Histogram-Based Splitting:** Identifies best splits via histogram bins for efficient memory use.

5. **Leaf-Wise Tree Growth:** Builds deeper trees by selecting leaf splits with maximum reduction.

6. **Gradient-Based One-Side Sampling (GOSS):** Prioritizes instances with larger gradients for faster convergence.

7. **Regularization:** Supports L1 and L2 regularization for model stability.

8. **Model Combination:** Adds trees iteratively to correct errors of the ensemble.

9. **Final Prediction:** Combines predictions for class labels (hard/soft voting).

10. **Hyperparameter Tuning:** Offers tunable parameters for optimizing performance.

11. **Versatility:** Applicable to various tasks - classification, regression, ranking, and more.

## **<font color='blue'>CatBoost**

CatBoost is an advanced gradient boosting library developed by Yandex, designed to excel in handling categorical features and delivering high-performance machine learning models. The name "CatBoost" stands for "Categorical Boosting," highlighting its unique ability to seamlessly incorporate categorical variables into the boosting process.

**Key Features**

1. **Categorical Feature Handling:**
  - CatBoost intelligently handles categorical variables without requiring extensive preprocessing like one-hot encoding.
  - It uses a novel algorithm that efficiently utilizes categorical information during model training.

2. **Ordered Boosting:**
   - CatBoost employs an ordered boosting strategy that optimizes the combination of features, enhancing the model's ability to capture relevant patterns.

3. **Symmetric Trees:**
  - The library utilizes symmetric trees to reduce memory usage during training, resulting in efficient and faster model building.

4. **Optimized Loss Functions:**
   - CatBoost provides specialized loss functions tailored for various tasks, such as classification, regression, and ranking.

5. **Regularization and Cross-Validation:**
   - CatBoost supports L2 regularization and offers cross-validation to prevent overfitting and ensure model robustness.

6. **Hyperparameter Optimization:**
  - A wide range of hyperparameters are available for tuning to optimize model performance.

7. **Automatic Handling of Missing Values:**
   - CatBoost automatically manages missing values during training, reducing the need for imputation.

### Missing-Value Treatment

* We will use median to impute missing values in Income column. We have decided to impute this variable with median as data is skewed.

##  Building the model

**First, let's create two functions to calculate different metrics and confusion matrix so that we don't have to use the same code repeatedly for each model.**

The following model performance classification function will return the accuracy, recall, precision and F1 score.

```python
# defining a function to compute different metrics to check performance of a classification model built using sklearn
def model_performance_classification_sklearn(model, predictors, target):
    """
    Function to compute different metrics to check classification model performance

    model: classifier
    predictors: independent variables
    target: dependent variable
    """

    # predicting using the independent variables
    pred = model.predict(predictors)

    acc = accuracy_score(target, pred)  # to compute Accuracy
    recall = recall_score(target, pred)  # to compute Recall
    precision = precision_score(target, pred)  # to compute Precision
    f1 = f1_score(target, pred)  # to compute F1-score

    # creating a dataframe of metrics
    df_perf = pd.DataFrame(
        {"Accuracy": acc, "Recall": recall, "Precision": precision, "F1": f1,},
        index=[0],
    )

    return df_perf
```

The following function will return the confusion matrix for a model

```python
def confusion_matrix_sklearn(model, predictors, target):
    """
    To plot the confusion_matrix with percentages

    model: classifier
    predictors: independent variables
    target: dependent variable
    """
    y_pred = model.predict(predictors)
    cm = confusion_matrix(target, y_pred)
    labels = np.asarray(
        [
            ["{0:0.0f}".format(item) + "\n{0:.2%}".format(item / cm.flatten().sum())]
            for item in cm.flatten()
        ]
    ).reshape(2, 2)

    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=labels, fmt="")
    plt.ylabel("True label")
    plt.xlabel("Predicted label")
```

### **LightGBM Classifier**

```python
# Create a LightGBM classifier
lgbm_classifier = LGBMClassifier(random_state=30)

# Train the LightGBM classifier
lgbm_classifier.fit(X_train, y_train)
```

```python
# Calculating different metrics on train and test set
lgbm_train = model_performance_classification_sklearn(
    lgbm_classifier, X_train, y_train
)

lgbm_test = model_performance_classification_sklearn(
    lgbm_classifier, X_test, y_test
)
print("Training performance:")
print(lgbm_train)

print("*************************************")

print("Test performance:")
print(lgbm_test)

print("*************************************")

# creating confusion matrix
confusion_matrix_sklearn(lgbm_classifier, X_train, y_train)
```
